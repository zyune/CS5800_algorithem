```python
import pulp
MyProbLP = pulp.LpProblem("LPProbDemo1", sense=pulp.LpMinimize)
tomato = pulp.LpVariable('tomato', lowBound=0, upBound=100, cat='Integer') 
lettuce = pulp.LpVariable('lettuce', lowBound=0, upBound=100, cat='Integer')
spinach = pulp.LpVariable('spinach', lowBound=0, upBound=100, cat='Integer') 
carrot = pulp.LpVariable('carrot', lowBound=0, upBound=100, cat='Integer')
oil = pulp.LpVariable('oil', lowBound=0, upBound=100, cat='Integer') 
MyProbLP += 21*tomato + 16*lettuce + 371*spinach + 346*carrot+ 884*oil  	# 设置目标函数
```

```python
MyProbLP += (tomato+lettuce+spinach+carrot+oil == 100)  # 等式约束  
MyProbLP += (0.85*tomato+1.62*lettuce+12.78*spinach+8.39*carrot >= 15)  # 不等式约束
# MyProbLP += (0.33*tomato+0.2*lettuce+1.58*spinach+1.39*carrot+100*oil >=2 )
MyProbLP += (2<= 0.33*tomato+0.2*lettuce+1.58*spinach+1.39*carrot+100*oil <= 6)
MyProbLP += (4.64*tomato+2.37*lettuce+74.69*spinach+80*carrot >= 4)
MyProbLP += (9*tomato+8*lettuce+7*spinach+508.20*carrot <= 100)
MyProbLP += (lettuce+spinach <= 50)
```

```python
MyProbLP.solve()
print("Status:", pulp.LpStatus[MyProbLP.status]) # 输出求解状态
for v in MyProbLP.variables():
    print(v.name, "=", v.varValue)  # 输出每个变量的最优值
print("F(x) = ", pulp.value(MyProbLP.objective))  #输出最优解的目标函数值    
```
